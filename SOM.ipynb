```python
import numpy as np
from minisom import MiniSom
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()
data = iris.data

# Initialize the SOM (e.g., a 10x10 grid)
som = MiniSom(x=10, y=10, input_len=data.shape[1], sigma=1.0, learning_rate=0.5)
som.random_weights_init(data)

# Train the SOM
som.train_random(data, num_iteration=100)

# Plotting the SOM
plt.figure(figsize=(10, 10))
for i, x in enumerate(data):
    winner = som.winner(x)  # Find the winning neuron
    plt.text(winner[0], winner[1], str(iris.target[i]), color='r', fontdict={'size': 12})

plt.title('SOM Visualization of Iris Dataset')
plt.show()
```
